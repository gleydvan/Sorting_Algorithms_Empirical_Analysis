```python
# Importar as bibliotecas necessárias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Atribuir o nome da planilha para a variável "arquivo"
tempos_nao_decrescentes = pd.read_csv('tempos_vector_nao_decrescentes(50 mil elementos).csv')
tempos_nao_crescentes = pd.read_csv('tempos_vector_nao_crescentes(50 mil elementos).csv')
tempos_aleatorios = pd.read_csv('tempos_vector_aleatorio(50 mil elementos).csv')
# Carregar a base de dados para um objeto pandas.io.excel.ExcelFile

tempos_nao_decrescentes
```

```python
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(13,13))
fig.tight_layout()
funcoes = ['Insertion','Selection','Bubble','Shell','Merge','Quick','Radix']

for i in range(0,7):
    ax=tempos_nao_decrescentes.plot(kind='line',x='tamanho_vetor',y=funcoes[i],
                   title=funcoes[i],
                   color='blue', legend=False,
                   ax=axes[i//3, i%3])
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    ax.set_xlabel("Quantidade de Elementos")
    ax.set_ylabel("Tempo em Milisegundos")
plt.tight_layout()
fig.savefig('./Não_Decrescente.png')
```

```python
fig2, axes2 = plt.subplots(nrows=3, ncols=1, figsize=(13,13))

funcoes = ['Insertion','Selection','Bubble','Shell','Merge','Quick','Radix']
funcoes_L = ['Selection','Quick']
funcoes_R = ['Insertion','Bubble','Shell','Merge','Radix']

ax = tempos_nao_decrescentes.plot(kind='line',x='tamanho_vetor',y=funcoes_L,title="Elementos em Ordem Não Decrescente",
                color=['green','red'], legend=True,ax=axes2[0])
ax.set_xlabel("Quantidade de Elementos")
ax.set_ylabel("Tempo em Milisegundos")

ax = tempos_nao_decrescentes.plot(kind='line',x='tamanho_vetor',y=funcoes_R,title="Elementos em Ordem Não Decrescente",
                color=['blue','pink','yellow','orange','black'], legend=True,ax=axes2[1])
ax.set_xlabel("Quantidade de Elementos")
ax.set_ylabel("Tempo em Milisegundos")

ax = tempos_nao_decrescentes.plot(kind='line',x='tamanho_vetor',y=[funcoes[0],funcoes[2]],title="Comparação entre Bubble sort e Insertion sort",
                color=['blue','pink'], legend=True,ax=axes2[2])
ax.set_xlabel("Quantidade de Elementos")
ax.set_ylabel("Tempo em Milisegundos")

plt.tight_layout()
fig2.savefig('./Não_Decrescente_Comparações.png')
```
